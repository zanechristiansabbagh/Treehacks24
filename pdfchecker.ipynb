```python
import PyPDF2
import openai

# Define the path to your PDF file
pdf_path = './TestSlides.pdf'

# Open the PDF file
with open(pdf_path, 'rb') as file:
    pdf = PyPDF2.PdfReader(file)
    number_of_pages =len(pdf.pages)
    print("number of pages: ", number_of_pages)
    
    # Iterate through all the pages and extract text
    pdf_texts = []
    for page_num in range(number_of_pages):
        page = pdf.pages[page_num]
        pdf_texts.append(page.extract_text())


# Assuming the OpenAI API key and necessary imports are already set up earlier in your notebook
# Function to create embeddings for the PDF text using OpenAI with RAG (Retrieval-Augmented Generation)
def create_pdf_embeddings_rag(pdf_texts):
    embeddings = []
    for text in pdf_texts:
        response = openai.Embedding.create(
            input=text,
            model="text-embedding-ada-002",  # An appropriate model for RAG tasks
            api_key="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"  # Place your OpenAI API key here
        )
        embeddings.append(response['data'][0]['embedding'])
    return embeddings

# Example usage
pdf_embeddings_rag = create_pdf_embeddings_rag(pdf_texts)
print(pdf_embeddings_rag)
```
